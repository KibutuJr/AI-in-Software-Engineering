```python
#manual generation 
def sort_dicts_manual(lst, key):
    n = len(lst)
    for i in range(n):
        for j in range(0, n-i-1):
            if lst[j][key] > lst[j+1][key]:
                lst[j], lst[j+1] = lst[j+1], lst[j]
    return lst
```

```python
# AI-Generated Implementation 
#implement the Python function sort_dicts_manual(lst, key)
def sort_dicts_ai(lst, key):
    # Bubble sort implementation for list of dicts by given key
    n = len(lst)
    for i in range(n):
        for j in range(0, n - i - 1):
            if lst[j][key] > lst[j + 1][key]:
                lst[j], lst[j + 1] = lst[j + 1], lst[j]
    return lst
```

```python
# 10,000 Random Dictionaries
import random 
import string

random_dicts = []  # initialize an empty list to hold the dictionaries
for _ in range(10000): # loop 10,000 times to create 10,000 dictionaries
    d = {
        'id': random.randint(1, 10000), # random int between 1 and 100000
        'name': ''.join(random.choices(string.ascii_letters, k=8)), # random 8 letter string
        'value': random.uniform(0, 1000) # random float btn 0 and 1000
    }
    random_dicts.append(d) # add the generated dictionary to the list

random_dicts
```

```python
%%timeit
sort_dicts_manual(random_dicts, 'value')
```

```python
%%timeit
sort_dicts_ai(random_dicts, 'value')
```
